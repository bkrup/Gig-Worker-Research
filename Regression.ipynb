```python
df['HS'] = np.where(df['EDUC'] < 73, 1, 0)
df['College'] = np.where(df['EDUC'].isin([73,81]), 1, 0)
df['Assoc'] = np.where(df['EDUC'].isin([91,92]), 1, 0)
df['Bach'] = np.where(df['EDUC'] == 111, 1, 0)
df['Grad'] = np.where((df['EDUC'] > 122) & (df['EDUC'] != 999), 1, 0)
df['Metro'] = np.where(df['METRO'].isin([2,3,4]), 1, 0)
df['White'] = np.where(df['RACE'] == 100, 1, 0)
df['Black'] = np.where(df['RACE'] == 200, 1, 0)
df['Asian'] = np.where(df['RACE'].isin([651,652]), 1, 0)
df['Hispanic'] = np.where(df['HISPAN'] != 0, 1, 0)
df['Citizen'] = np.where(df['CITIZEN'].isin([1,2,3]), 1, 0)
df['Naturalized'] = np.where(df['CITIZEN'] == 4, 1, 0)
df['Illegal'] = np.where(df['CITIZEN'] == 5, 1, 0)
df['Sep/Div/Wid'] = np.where(df['MARST'].isin([3,4,5]), 1, 0)
df['Married'] = np.where(df['MARST'].isin([1,2]), 1, 0)
df['Single'] = np.where(df['MARST'] == 6, 1, 0)
df['Mining'] = np.where((df['IND'] >= 290) & (df['IND'] <= 490), 1, 0)
df['Construction'] = np.where((df['IND'] == 770), 1, 0)
df['Manufacturing'] = np.where((df['IND'] >= 1070) & (df['IND'] <= 3990), 1, 0)
df['Wholesaler'] = np.where((df['IND'] >= 4070) & (df['IND'] <= 4590), 1, 0)
df['Retailers'] = np.where((df['IND'] >= 4670) & (df['IND'] <= 5790), 1, 0)
df['Transportation'] = np.where((df['IND'] >= 6070) & (df['IND'] <= 6290), 1, 0)
df['Courier'] = np.where((df['IND'] >= 6370) & (df['IND'] <= 6380), 1, 0)
df['Utilities'] = np.where((df['IND'] >= 570) & (df['IND'] <= 690), 1, 0)
df['Communications'] = np.where((df['IND'] >= 6470) & (df['IND'] <= 6780), 1, 0)
df['Banking'] = np.where((df['IND'] >= 6870) & (df['IND'] <= 7070), 1, 0)
df['Rentals'] = np.where((df['IND'] >= 7080) & (df['IND'] <= 7190), 1, 0)
df['Services'] = np.where((df['IND'] >= 7270) & (df['IND'] <= 7490), 1, 0)
df['Support'] = np.where((df['IND'] >= 7570) & (df['IND'] <= 7790), 1, 0)
df['Edu_Ind'] = np.where((df['IND'] >= 7860) & (df['IND'] <= 7890), 1, 0)
df['Healthcare_Ind'] = np.where((df['IND'] >= 7970) & (df['IND'] <= 8190), 1, 0)
df['Nursing'] = np.where((df['IND'] >= 8270) & (df['IND'] <= 8470), 1, 0)
df['Entertainment'] = np.where((df['IND'] >= 8560) & (df['IND'] <= 8590), 1, 0)
df['Travel'] = np.where((df['IND'] >= 8660) & (df['IND'] <= 8690), 1, 0)
df['Personal_Ind'] = np.where((df['IND'] >= 8770) & (df['IND'] <= 9090) & (df['IND'] == 9290), 1, 0)
df['Religious'] = np.where((df['IND'] >= 9160) & (df['IND'] <= 9190), 1, 0)
df['Public'] = np.where((df['IND'] >= 9370) & (df['IND'] <= 9590), 1, 0)
df['Management_Occ'] = np.where((df['OCC'] >= 10) & (df['OCC'] <= 430), 1, 0)
df['Business_Occ'] = np.where((df['OCC'] >= 500) & (df['OCC'] <= 740), 1, 0)
df['Financial_Occ'] = np.where((df['OCC'] >= 800) & (df['OCC'] <= 950), 1, 0)
df['Math_Occ'] = np.where((df['OCC'] >= 1005) & (df['OCC'] <= 1240), 1, 0)
df['Architecture_Occ'] = np.where((df['OCC'] >= 1300) & (df['OCC'] <= 1560), 1, 0)
df['Life_Occ'] = np.where((df['OCC'] >= 1600) & (df['OCC'] <= 1965), 1, 0)
df['Community_Occ'] = np.where((df['OCC'] >= 2000) & (df['OCC'] <= 2060), 1, 0)
df['Legal_Occ'] = np.where((df['OCC'] >= 2100) & (df['OCC'] <= 2160), 1, 0)
df['Edu_Occ'] = np.where((df['OCC'] >= 2200) & (df['OCC'] <= 2550), 1, 0)
df['Arts_Occ'] = np.where((df['OCC'] >= 2600) & (df['OCC'] <= 2920), 1, 0)
df['Healthcare_Occ'] = np.where((df['OCC'] >= 3000) & (df['OCC'] <= 3540), 1, 0)
df['Health_Support_Occ'] = np.where((df['OCC'] >= 3600) & (df['OCC'] <= 3655), 1, 0)
df['Protective_Service_Occ'] = np.where((df['OCC'] >= 3700) & (df['OCC'] <= 3955), 1, 0)
df['Food_Occ'] = np.where((df['OCC'] >= 4000) & (df['OCC'] <= 4150), 1, 0)
df['Cleaning_Occ'] = np.where((df['OCC'] >= 4200) & (df['OCC'] <= 4250), 1, 0)
df['Personal_Care_Occ'] = np.where((df['OCC'] >= 4300) & (df['OCC'] <= 4650), 1, 0)
df['Sales_Occ'] = np.where((df['OCC'] >= 4700) & (df['OCC'] <= 4965), 1, 0)
df['Office_Support_Occ'] = np.where((df['OCC'] >= 5000) & (df['OCC'] <= 5940), 1, 0)
df['Construction_Occ'] = np.where((df['OCC'] >= 6200) & (df['OCC'] <= 6765), 1, 0)
df['Extraction_Occ'] = np.where((df['OCC'] >= 6800) & (df['OCC'] <= 6940), 1, 0)
df['Inst_Maint_Repair_Occ'] = np.where((df['OCC'] >= 7000) & (df['OCC'] <= 7630), 1, 0)
df['Production_Occ'] = np.where((df['OCC'] >= 7700) & (df['OCC'] <= 8965), 1, 0)
df['Transportation_Occ'] = np.where((df['OCC'] >= 9000) & (df['OCC'] <= 9750), 1, 0)
df['Employed'] = np.where(df['EMPSTAT'].isin([10]), 1, 0)
df['Unemployed'] = np.where(df['EMPSTAT'].isin([12,21,22]), 1, 0)
df['NILF'] = np.where(df['EMPSTAT'] > 30, 1, 0)
df['Change'] = df['AHRSWORKT'] - df['UHRSWORKT']
```
